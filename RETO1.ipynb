```python
kilos_floculante = int(input())
kilos_cal = int(2*kilos_floculante+4)
litros_cloro = int((kilos_cal+kilos_floculante)/5)
print(str(kilos_floculante)+" "+str(kilos_cal)+" "+str(litros_cloro))
if litros_cloro <=20:
  print("uno")
elif litros_cloro>20 and litros_cloro<=30:
  print("dos")
elif litros_cloro>30 and litros_cloro<=50:
  print("tres")
else:
  print("cuatro")

```
