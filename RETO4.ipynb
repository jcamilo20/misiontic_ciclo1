Los proveedores de equipo de cmputo, normalmente identifican los bienes que tienen para la venta en un catlogo mediante parejas <cdigo_estandar:valor> en formato JSON.

Una empresa va a adquirir varios equipos de cmputo de varias marcas (servidores, estaciones de trabajo, computadores de escritorio, porttiles, tabletas, enrutadores, switches, Access points), para todas sus sedes. Y hace una licitacin publicando una relacin de los equipos que requiere, identificados mediante de cdigos estndar de una letra.

Uno de los proveedores de equipo requiere un programa para cruzar la lista de equipos solicitados por licitacin contra su propio catalogo ya que quiere saber cules equipos puede ofrecer y cul sera el valor total de la oferta, si de cada equipo se compra una unidad.

Entrada:

Lnea1: parejas <cdigo_estandar:valor> en formato JSON del catlogo del proveedor y linea2: relacin total de letras que identifican los equipos solicitados por licitacin. En este caso los equipos se identifican por letras del alfabeto, separados por espacio.

Salida:

Lnea 1: valor total de los equipos solicitados y que el proveedor puede ofrecer, representado por un nmero entero.

linea2: relacin de cdigos que identifican los equipos que el proveedor puede ofrecer y que son solicitados en la licitacin.

```python
def venta_cruzada():
  import json
  catalogo_proveedor = input()
  catalogo_proveedor = json.loads(catalogo_proveedor)  #convierte el string del input a formato json que entiende python (diccionario)
  pedido = input().split(" ")
  contador=0
  match=[]
  for i in pedido:
    for clave, valor in catalogo_proveedor.items():
      if i == clave:
        contador=contador+valor
        match.append(clave)
  print(contador)
  print(*match)
  
venta_cruzada()
  


```
