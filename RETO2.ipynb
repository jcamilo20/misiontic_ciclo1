En la asamblea ordinaria de un conjunto residencial, van a elegir el consejo de administracin, a partir de dos planchas: plancha 1 y plancha 2; cada plancha va a tener n candidatos, que se identifican por una letra nica. Ud va a elaborar el programa para el escrutinio. Los votos solo contienen una letra que identifica a uno de los candidatos, cualquiera sea su plancha, recuerde puede haber votos en blanco tambin. El programa va a incrementar un contador cada que se procese un voto por la plancha 1 y va a incrementar su propio contador para la plancha 2, excepto que sea un voto en blanco.

Entradas:

lnea 1: serie de las letras que identifican los integrantes de la plancha 1.

Lnea 2: algo similar para la plancha 2.

Lnea 3: serie de muchas letras, una por cada voto.

Las letras se ingresan juntas, es decir sin espacios entre ellas

 

Salida:

una de las letras P, N, I, se imprime cada vez que se lee pblicamente el voto. Imprime la letra P si la plancha 1 tiene en ese momento mas votos que la plancha 2; Imprime N si la plancha 2 tiene en ese momento mas votos que la plancha 1; imprime la letra I, si en ese momento hay empate.

```python
def audita(plan1,plan2,votos):
  cuenta_votos1=0
  cuenta_votos2=0
  for i in votos:
    if i in plan1:
        cuenta_votos1+=1
    if i in plan2:
        cuenta_votos2+=1
    if cuenta_votos1>cuenta_votos2:
        print("P", end="")
    elif cuenta_votos2>cuenta_votos1:
        print("N", end="" )
    else:
        print("I", end="")

plancha1 = input()
plancha2 = input()
votacion = input()

audita(plancha1,plancha2,votacion)
```
